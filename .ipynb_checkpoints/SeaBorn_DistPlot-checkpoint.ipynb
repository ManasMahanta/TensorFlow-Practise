```python
import numpy as np
```

```python
import seaborn as sns
```

```python
dist=np.random.randn(150)
dist
```

```python
sns.distplot(dist,color='red')
```

```python
import pandas as pd
```

```python
label_dist=pd.Series(dist,name='Variable x')
```

```python
sns.distplot(label_dist,color='blue')
```

```python
sns.distplot(label_dist,color='blue',vertical=True)
```

```python
sns.distplot(label_dist,color='blue',rug=True,hist=False)
```
