```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='sex',palette='winter_r',order=['Sat','Sun','Fri','Thur'])
```

```python
from numpy import median
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.median)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.median,ci=66)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.median,capsize=0.3)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.median,capsize=0.9)
```

```python
sns.barplot(x='size',y='tip',data=tips,estimator=np.median,color='red',saturation=0.9)
```
